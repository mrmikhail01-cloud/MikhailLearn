"Уровень 3"

```python
while True:
    password = input("Придумайте пароль: ")
    
    if len(password) >= 8 and any(c.islower() for c in password) and any(c.isupper() for c in password):
        print("Пароль принят!")
        break
    else:
        print("Пароль должен быть длиннее 8 символов и содержать как строчные, так и заглавные буквы.")
```
